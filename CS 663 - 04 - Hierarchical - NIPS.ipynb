```python
import pandas as pd
df = pd.read_csv('NIPS_1987-2015.csv')
df
```

```python
new_df = df[df.columns[1:]]
new_df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(new_df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(15, 10))
dendrogram(Z)
plt.show()
```
